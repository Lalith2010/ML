# PROJECT - 1 "CLASSIFICATION"

CLASSIFICATION is a technique where the data will be categorized into a given number of classes. The main goal of a classification problem is to identify the category/class to which a new data will fall under

PROJECT DESCRIPTION AND AGENDA
----------------------------------------------
Mushroom hunting (otherwise known as "Shrooming") is enjoying new peaks in popularity. The concept behind this handpick is simple and straight forward which helps us to learn which features of a particular mushroom spell certain death and which are most palatable in this dataset of mushroom characteristics. 
                                                                                        Talking about the dataset, this dataset includes recordings of hypothetical samples corresponding to 23 species of gilled mushrooms in the regions of Agaricus and Lepiota Family Mushroom drawn from The Audubon Society Field Guide to North American Mushrooms (1981). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom
                                                                                        
THE AGENDA OF THIS PROJECT IS:

*To identify what types of machine learning models perform best on this dataset?

*Which features are most indicative of a poisonous mushroom?

>CHECK FOR NULL VALUES

The very immediate step after loading the data would be checking for any missing values that are present in the data. Missing data can reduce the statistical power of a study and can produce biased estimates, leading to invalid conclusions. Hence, it is important to remove any missing values that are present in the dataset

```python
for col in df.columns:
    print("{} : {}".format(col, df[col].isnull().sum()))
```

As we see, there are missing values present in 8 columns in the dataset. Missing data can be handled in multiple ways: 
1. By imputing the values with simple average or median
2. Replace by most frequent value
3. By simply droping the rows that comtain the missing values.
The type of handling method depends mainly on type of data, as it is not appropriate to perform statistical calculations on categorical data. Besides, due to the presence of sufficient amount of data and by considering the type of data that we have, in this project we are going to drop the rowns that contains those missing values.

```python
def plot_data(hue, data):
    for i, col in enumerate(data.columns):
        plt.figure(i)
        sns.set(rc={'figure.figsize':(12,9)})
        ax = sns.countplot(x=data[col], hue=hue, data=data) 
```

Data Wrangling
------------------

It is not recommended to apply label encoding for those features that have values more than 2. The reason for this is to avoid giving more importance to the other values as the Label ecoding assigns values as 0,1,2,..

In order to handle this type of problems, we use One hot vectorization

```python
X_enc = pd.get_dummies(X)
X_enc.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X_enc)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_enc = le.fit_transform(y.values.ravel())

y_enc
```

Model Building
------------------
Since, classification comes under supervised learning, it is important to divide the data into Training and Testing data sets before building any model.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std,y_enc,train_size=0.75,test_size=0.25,random_state=4)
```

ENSEMBLING
-----------------
>HARD VOTING

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import VotingClassifier

log_clf = LogisticRegression()
log_clf.fit(X_train, y_train)
knn_clf = KNeighborsClassifier(7)
knn_clf.fit(X_train, y_train)


voting_clf = VotingClassifier(estimators=[('lr', log_clf), ('knn', knn_clf)], voting='hard')
voting_clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
for clf in (log_clf, knn_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

>SOFT VOTING

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import VotingClassifier

log_clf = LogisticRegression()
log_clf.fit(X_train, y_train)
knn_clf = KNeighborsClassifier(7)
knn_clf.fit(X_train, y_train)


voting_clf = VotingClassifier(estimators=[('lr', log_clf), ('knn', knn_clf)], voting='soft')
voting_clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
for clf in (log_clf, knn_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```
